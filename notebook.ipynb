### **Data Cleaning**

```python
# Finding all missing values
df.isnull().sum()

# Dropping all missing value rows
df.dropna(inplace=True)

df.isnull().sum()
```

```python
df.shape
```

```python
# Handling Duplicates
df.duplicated().sum()
```

```python
# Deleting all duplicate rows
df.drop_duplicates(inplace=True)

df.duplicated().sum()
```

```python
df.dtypes
```

```python
# Handling data types

df['id'] = df['id'].astype(object)

df['host_id'] = df['host_id'].astype(object)

df.dtypes
```

### **Exploratory Data Analysis**

#### **Univariate Analysis**

```python
# Price Distribution

sns.histplot(data=df, x='price')
```

```python
# Identifying Outliers in price

sns.boxplot(data=df, x='price')
```

```python
# Handling outlier in price

df2 = df[df['price'] < 1500]
```

```python
sns.boxplot(data=df2, x='price')
```

```python
# Price Distribution
plt.figure(figsize=(8,5))
plt.title('Price Distribution')
sns.histplot(data=df2, x='price', bins=100)
plt.ylabel("Frequency")
plt.show()
```

```python
# 365 days Availability 

# Price Distribution
plt.figure(figsize=(6,3))
plt.title('365 Days Availability')
sns.histplot(data=df2, x='availability_365')
plt.ylabel("Frequency")
plt.show()
```

```python
df2.groupby(by='neighbourhood_group')['price'].mean()
```

### **Feature Engineering**

```python
# price per bed
df2 = df.copy()
df2.loc[:, 'price_per_bed'] = df2['price'] / df2['beds']
df2.head()
```

```python
# average price per bed 
df2.groupby(by='neighbourhood_group')['price_per_bed'].mean()
```

#### **Bi-variate Analysis**

```python
df2.columns
```

```python
# Price v/s neighbourhood relation
plt.figure(figsize=(8,5))
plt.title('Price v/s Neighbourhood Relation')
sns.barplot(data=df2, x='neighbourhood_group', y='price', hue='room_type')
```

```python
# No. of reviews v/s price relation
plt.figure(figsize=(8,5))
plt.title('Number of Reviews v/s Price Relation')
sns.scatterplot (data=df2, x='number_of_reviews', y='price', hue='neighbourhood_group')
```

```python
df2.dtypes
```

```python
sns.pairplot(data=df2, vars=['price', 'minimum_nights', 'number_of_reviews', 'availability_365'], hue='room_type')
```

```python
# Geographical Distribution of Listings
plt.figure(figsize=(10,7))
plt.title('Geographical Distribution of the Listings')
sns.scatterplot(data=df2, x='longitude', y='latitude', hue='room_type')
plt.show()
```

```python
df2.dtypes
```

```python
# Correlation
corr = df2[['latitude', 'longitude', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'availability_365', 'beds']].corr()
corr

plt.figure(figsize=(8,8))
sns.heatmap(corr, annot=True)
plt.show()
```
